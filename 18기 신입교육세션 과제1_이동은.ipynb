과제1

데이터명: 대한체육회_스포츠지원포털 팀등록정보
데이터 특징
- 출처: 공공데이터포털(https://www.data.go.kr/data/15052697/fileData.do)
- 분야: 스포츠
- 내용: 이 데이터는 대한체육회 스포츠지원포털에 등록된 스포츠팀 정보 관련 데이터입니다. 각 팀별로 팀명, 등록년도, 종목, 세부종목, 소속, 시도 정보 등이 포함되어있습니다. 정형 데이터(CSV)로 분석/시각화가 쉽고, 스포츠 종목/시도별 비교 분석이 가능합니다. 스포츠 산업 및 팀 구조 이해에 직접 활용 가능한 데이터이고 팀 분포, 종목별 규모나 지역별 분포 같은 EDA 과제에 적합해 이 데이터를 고르게 되었습니다.



데이터명: Bank Marketing Dataset
데이터 특징
- 출처: kaggle (https://www.kaggle.com/datasets/janiobachmann/bank-marketing-dataset)
- 분야: 마케팅, 금융
- 내용: 이 데이터는 은행의 마케팅 캠페인 관련 데이터입니다. 고객 특성, 마케팅 캠페인 접촉 정보, 결과(가입 여부) 등이 포함된 데이터로, 고객 행동 분석이나 예측 모델링에 적합합니다. 고객 연령, 직업, 결혼 상태 등 설명 변수가 다양합니다. 또한 캠페인 결과(가입/비가입)로 이진 분류 모델을 만들기에 적합한 데이터입니다. 금융기관의 마케팅 전략과 고객 반응 데이터를 동시에 갖춘 실전 비즈니스 데이터이기 때문에 마케팅/금융 분석 과제에 좋아서 이 데이터 고르게 되었습니다.
