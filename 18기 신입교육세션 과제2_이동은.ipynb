나이 변수 결측치 처리

이름 변수의 Mr, Miss, Mrs, Master, Dr을 기준으로 각 그룹의 평균 나이를 계산하고 그 평균 나이로 결측치를 처리한다

```python
data2["Name"]
```

```python
data2["NameSep"]=data2["Name"].str.split(",").str[1].str.split(".").str[0].str.strip()
```

```python
data2["NameSep"]
```

```python
data2["NameSep"]=data2["NameSep"].replace({
    "Mlle":"Miss",
    "Ms":"Miss",
    
    "Mme":"Mrs",

    "Rev":"Dr",
    "Major":"Dr",
    "Col":"Dr",
    "Capt":"Dr",

    "Don":"Other",
    "Sir":"Other",
    "Lady":"Other",
    "Jonkheer":"Other",
    "the Countess":"Other"
})
```

```python
data2["NameSep"]
```

```python
meanAge=data2.groupby("NameSep")["Age"].mean()
print(meanAge)
```

```python
def fill_age(row):
    if pd.isnull(row['Age']):
        return meanAge[row['NameSep']]
    else:
        return row['Age']

data2['Age'] = data2.apply(fill_age, axis=1)
```
